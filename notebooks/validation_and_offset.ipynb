```python
import json, sys
sys.path.append('../')
from pymatgen.cli.pmg_query import MPRester
mprester = MPRester('mlcC4gtXFVqN9WLv')
from scipy.stats import linregress
from matplotlib import pylab as plt
import statsmodels.api as sm 

from constants import Eads_ref_oc20, Eads_ref_rpbe
from constants import Ncorr_db as Ncorr
from constants import Ocorr_db as Ocorr
```

```python
validation_dft = json.load(open('../datasets/DFT_test_set.json', 'rb'))
validation_cgcnn_relaxed = json.load(open('../datasets/ML_test_set.json', 'rb'))
```

```python
# Test ML with DFT calculated data from this work and Liu et al.

count = 0
plt.figure(figsize=(5.25,5))
plt.rcParams['font.family'] = ['serif']

x, y = [], []
xn, xo, yn, yo = [], [], [], []
distinct = []
for k in validation_dft.keys():
    if k not in validation_cgcnn_relaxed.keys():
        continue
    formula = mprester.get_entry_by_material_id(k).composition.reduced_formula if 'mp-' in k else k
    if 'old' in k:
        formula = k.split('_')[0]
    
    if any([el in formula for el in ['La', 'Sn']]):
        continue
        
    for hkl in validation_cgcnn_relaxed[k].keys():
        for ads in validation_cgcnn_relaxed[k][hkl].keys():
            c = 'b' if ads == 'N' else 'r'
            m = '^' if 'liu' in k else 'o'
            for i, ml_eads in enumerate(validation_cgcnn_relaxed[k][hkl][ads]):
                Eads_cgcnn = ml_eads + Eads_ref_oc20['*%s' %(ads)] - Eads_ref_rpbe['*%s' %(ads)]
                eads_dft = validation_dft[k][hkl][ads][i]
                plt.scatter(Eads_cgcnn, eads_dft, c=c, edgecolor='k', s=100, marker=m)
                if m == '^':
                    count+=1
                if formula not in distinct and 'liu' not in formula:
                    distinct.append(formula)
                x.append(Eads_cgcnn)
                y.append(eads_dft)
                if ads == 'N':
                    xn.append(Eads_cgcnn) 
                    yn.append(eads_dft)
                else:
                    xo.append(Eads_cgcnn)  
                    yo.append(eads_dft)

slope, intercept, r, p, se = linregress(x, y)
corr = [intercept, slope]
plt.plot([-100, 100], [-100*slope+intercept, 100*slope+intercept], 'g--',
         label=r'$E_{DFT}=E_{ML}%.2f$, $R^2=%.2f$' %(intercept, r**2))
mae = sm.tools.eval_measures.meanabs(x, y)
print('Everything',slope, intercept, r**2, p, se)
print('MAE', mae)

slope, intercept, r, p, se = linregress(xn, yn)
Ncorr = [intercept, slope]
plt.plot([-100, 100], [-100*slope+intercept, 100*slope+intercept], 'b--',
         label=r'$E^{*N}_{DFT}=E^{*N}_{ML}%.2f$, $R^2=%.2f$' %(intercept, r**2))
mae = sm.tools.eval_measures.meanabs(xn, yn)
print("N", slope, intercept, r**2, se)
print('MAE', mae)

slope, intercept, r, p, se = linregress(xo, yo)
Ocorr = [intercept, slope]
plt.plot([-100, 100], [-100*slope+intercept, 100*slope+intercept], 'r--', 
         label=r'$E^{*O}_{DFT}=E^{*O}_{ML}+%.2f$, $R^2=%.2f$' %(intercept, r**2))
mae = sm.tools.eval_measures.meanabs(xo, yo)
print("O", slope, intercept, r**2, se)
print('MAE', mae)
    
lims = [-9.565565836688224, -3.016821084124608]
plt.xlim(lims)
plt.ylim(lims)
plt.plot([-100,100], [-100,100], 'k--')
plt.ylabel(r'DFT adsorption energy (eV)', fontsize=15)
plt.xlabel('ML adsorption energy (eV)', fontsize=15)

plt.legend(fontsize=12.5, loc=2, labelspacing=0.05, 
           columnspacing=0.05, borderpad=0.09, 
           handletextpad=0.05)
plt.plot([-100,100], [-100, 100], 'k--')
plt.tight_layout()
plt.show()
plt.close()
```
